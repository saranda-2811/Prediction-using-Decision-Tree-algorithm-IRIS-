```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
%matplotlib inline
```

```python
df  = pd.read_csv(r'C:\Users\saran\Documents\datasets\Iris.csv')
```

```python
df.info()
```

```python
categories = df.Species.unique() 
categories
```

```python
df.Species.value_counts()
```

```python
df['Class']=''

for i in range(len(df.Species)):
    if df.Species[i] == categories[0]:
        df.Class[i] = 0
    elif df.Species[i] == categories[1]:
        df.Class[i] = 1
    elif df.Species[i] == categories[2]:
        df.Class[i] = 2

df
```

```python
x = pd.concat([df['SepalLengthCm'],df['SepalWidthCm'], df['PetalLengthCm'], df['PetalWidthCm']],axis=1)
y = df['Class']
```

```python
x
```

```python
y=y.astype('int')
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
from sklearn import tree

fn=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (15,10), dpi=300)
tree.plot_tree(model,
               feature_names = fn, 
               class_names=cn,
               filled = True); 
```

### TESTING THE MODEL

```python
# Making predictions
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)

# Calculate the accuracy
from sklearn.metrics import accuracy_score
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
print('The training accuracy is', train_accuracy)
print('The test accuracy is', test_accuracy)
```

```python
from sklearn.metrics import accuracy_score, make_scorer
from sklearn.model_selection import GridSearchCV

clf = DecisionTreeClassifier(random_state = 23)
parameters = {'max_depth': np.arange(3, 21), 'min_samples_leaf': np.arange(2, 11), 'min_samples_split': np.arange(4, 11)}
scorer = make_scorer(accuracy_score)
grid = GridSearchCV(estimator = clf, param_grid = parameters, scoring = scorer)


# Train the model
grid_fit = grid.fit(x_train, y_train)
best_clf = grid_fit.best_estimator_
print(best_clf)

# Make predictions
y_train_pred = best_clf.predict(x_train)
y_test_pred = best_clf.predict(x_test)

# Calculate the accuracy
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
print('The training accuracy is', train_accuracy)
print('The test accuracy is', test_accuracy)
```
